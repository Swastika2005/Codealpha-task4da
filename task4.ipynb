```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Product': ['Book', 'Laptop', 'Phone', 'Shoes', 'Camera'],
    'Review': [
        'Absolutely loved it! Very useful and well-written.',
        'Terrible performance, it kept crashing!',
        'Okay phone. Not the best, not the worst.',
        'So comfortable and fits perfectly!',
        'Poor quality images. Not worth the price.'
    ]
}
df = pd.DataFrame(data)
def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
df['Sentiment'] = df['Review'].apply(get_sentiment)
df['Polarity'] = df['Review'].apply(lambda x: TextBlob(x).sentiment.polarity)
print(df)
sns.countplot(data=df, x='Sentiment')
plt.title("Sentiment of Amazon Reviews")
plt.show()
```
